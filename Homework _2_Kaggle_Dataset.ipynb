```python


try:
    # Loaded dataset into a DataFrame and named it Video_Game_Data
    Video_Game_Data = pd.read_csv(r'C:\Users\JackT\CodeLancs\Andre\Datasets\archive.zip')

    # Identifying non-numerical columns
    non_numerical_cols = Video_Game_Data.select_dtypes(include=['object']).columns
    
    print(Video_Game_Data.head())

except Exception as e:
    # Printing an error message if an exception occurs
    print("An error occurred:", e)



#non_numerical_cols = Video_Game_Data.select_dtypes(include=['object']).columns
```

```python
#Rralised later that encodin was not the solution but was a useful method to learn.
label_encoder = LabelEncoder()
for col in non_numerical_cols:
    Video_Game_Data[col] = label_encoder.fit_transform(Video_Game_Data[col])
```

```python
import matplotlib.pyplot as plt

plt.scatter(Video_Game_Data['Age'], Video_Game_Data['favorite_game'])
plt.xlabel('Age')
plt.ylabel('favorite_game')
plt.title('Scatter Plot')
plt.show()
```

```python

# Attempting to use a bar chart with DataFrame containing non-integer values like peoples favorite video games

# Counted the frequency of each unique value in the 'Favorite_Game' column
game_counts = Video_Game_Data['favorite_game'].value_counts()

# Created a bar chart
plt.figure(figsize=(10, 6))
game_counts.plot(kind='bar')
plt.xlabel('Favorite Video Game')
plt.ylabel('Frequency')
plt.title('Favorite Video Games')
plt.xticks(rotation=45)  # Rotated x-axis labels for better visibility
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have a DataFrame named Video_Game_Data with a column 'Favorite_Game'
# containing non-encoded video game titles
Video_Game_Data = pd.read_csv(r'C:\Users\JackT\CodeLancs\Andre\Datasets\archive.zip')

# Count the frequency of each unique value in the 'Favorite_Game' column
game_counts = Video_Game_Data['favorite_game'].value_counts()

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(game_counts, labels=game_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Favorite Video Games')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

Video_Game_Data = pd.read_csv(r'C:\Users\JackT\CodeLancs\Andre\Datasets\archive.zip')

# Create a box plot or violin plot to compare age to favorite video game
plt.figure(figsize=(10, 6))
sns.boxplot(x='favorite_game', y='Age', data=Video_Game_Data)
plt.xlabel('Favorite Video Game')
plt.ylabel('Age')
plt.title('Age Distribution by Favorite Video Game')
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility
plt.show()
```
