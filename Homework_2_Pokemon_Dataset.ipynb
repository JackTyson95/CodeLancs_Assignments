```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
pokemon_df = pd.read_csv(r"C:\Users\JackT\CodeLancs\Andre\Datasets\Pokemon.csv", sep=',')
```

```python
pokemon_df.keys()
```

```python
pokemon_df.plot.scatter(x='Sp. Atk',y='Attack')
#Two keys compared against each other.
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10,6))

fig.suptitle('Stat comparisons')

pokemon_df.plot.scatter(x='Defense', y='Attack', c='black', s=1, ax=axes[0,0])
pokemon_df.plot.scatter(x='Speed',y='Attack', c='black', s=1, ax=axes[1,0])
pokemon_df.plot.scatter(x='HP',y='Attack',s=1, c='black', ax=axes[0,1])
pokemon_df.plot.scatter(x='Sp. Atk',y='Attack', c='black', s=1, ax=axes[1,1])
#Comparing attack versus various stats
```

```python

fig1, axes1 = plt.subplots(nrows=1, ncols=1, figsize=(8, 6))


for key in pokemon_df['Type 1'].unique():
    axes1.scatter(x=pokemon_df[pokemon_df['Type 1'] == key]['Attack'],
                  y=pokemon_df[pokemon_df['Type 1'] == key]['Defense'],
                  s=5,
                  label=key)
axes1.legend()  


fig2, axes2 = plt.subplots(nrows=1, ncols=1, figsize=(8, 6))


for key in pokemon_df:  
    axes2.scatter(x=pokemon_df[pokemon_df['Type 2'] == key]['Attack'],
                  y=pokemon_df[pokemon_df['Type 2'] == key]['Defense'],
                  s=5,
                  label=key)
axes2.legend()  

plt.show()



#Looking at the distribution of attack and defense by type.
```

```python



fig, (axes1, axes2) = plt.subplots(nrows=1, ncols=2, figsize=(16, 6))


for key in pokemon_df['Type 1'].unique():
    axes1.scatter(x=pokemon_df[pokemon_df['Type 1'] == key]['Attack'],
                  y=pokemon_df[pokemon_df['Type 1'] == key]['Defense'],
                  s=5,
                  label=key)
axes1.legend()
axes1.set_title('Type 1')  


for key in pokemon_df['Type 2'].unique():
    axes2.scatter(x=pokemon_df[pokemon_df['Type 2'] == key]['Attack'],
                  y=pokemon_df[pokemon_df['Type 2'] == key]['Speed'],
                  s=5,
                  label=key)
axes2.legend()  
axes2.set_title('Type 2')  

plt.show()


#Comparing attack and defense distribution versus attack and speed.


```
